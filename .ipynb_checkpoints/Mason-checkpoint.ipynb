# Best Neighborhoods in Pittsburgh

## Metric 
**Monthly Air Quality Report** \
This data represents the air quality for different neighborhoods in Pittsburgh, sourced from WPRDC. It reflects the number from the AQI index to show the health of the air occurring in each area. This measure helps identify which communities experience better air, thereby offering residents a health and quality of living.

```python
import pandas as pd
%matplotlib inline
```

## This is the code I used to read the data from the file and show the average AQI index for each of them

```python
# Read file
url = ""
data = pd.read_csv("AirQuality.csv")

# Combine the repeats into one so it doesn't mess with the final data
rename = {"Lawrenceville 2": "Lawrenceville","Liberty 2": "Liberty"}
data["site"] = data["site"].replace(rename)

# Creating the count variables
count_neighborhoods = {}
sum_quality = {}

# loop through and collect the data while putting it into the previously created count variables
for neighborhood, index_value in zip(data["site"], data["index_value"]):
    if neighborhood in count_neighborhoods:
        count_neighborhoods[neighborhood] += 1
        sum_quality[neighborhood] += index_value
    else:
        count_neighborhoods[neighborhood] = 1
        sum_quality[neighborhood] = index_value

# Create a variable to combine the neighborhood while computing and show the average
combined = {}
for neighborhood in count_neighborhoods:
    combined[neighborhood] = {
        "count": count_neighborhoods[neighborhood],
        "avg_index": sum_quality[neighborhood] / count_neighborhoods[neighborhood]
    }

# Sorting the combine variable to make it clean
sorted_combine = dict(sorted(combined.items(), key=lambda item: item[1]["avg_index"]))

# Printing
print(f"{'Neighborhood':<25} {'Reports':<10} {'Average Index':<12}")
print("-" * 50)
for neighborhood, values in sorted_combine.items():
    print(f"{neighborhood:<25} {values['count']:<10} {values['avg_index']:<12}")
```
