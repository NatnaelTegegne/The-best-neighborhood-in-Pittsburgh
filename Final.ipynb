# Best Neighborhoods in Pittsburgh

### Introduction

In this analysis, we aimed to identify the best neighborhood in Pittsburgh by comparing data across multiple metrics, including crime rate, income level, and air quality. To ensure consistency, we focused only on neighborhoods that appeared across all three datasets. After cleaning and aligning the data, we applied our evaluation metrics to determine which neighborhoods performed best overall.

After compiling and cleaning the datasets for crime, income, and air quality, we noticed that not all neighborhoods were consistently represented across the three sources. Some rows contained missing values (NaN)—especially in the crime dataset—indicating cases where the specific neighborhood where the crime occurred was not identified. To maintain accuracy and fairness in comparison, these unidentified records were excluded from our analysis.

Next, we focused on finding the common neighborhoods that existed across all three datasets. This ensured that every neighborhood considered had complete data for crime rate, average income, and air quality index.

Once the datasets were aligned, we normalized each metric for comparability:

**Crime Rate**: Lower values indicate safer neighborhoods.

**Average Income**: Higher values suggest better economic status.

**Air Quality Index (AQI)**: Lower values correspond to cleaner air.

We then ranked the neighborhoods within each metric to evaluate overall performance. Based on these rankings:

Lawrenceville ranked in the top 35% for crime rate (safer than most neighborhoods).

It also placed in the top 30% for income, indicating relatively strong economic conditions.

For air quality, Lawrenceville again ranked in the top 30%, showing better environmental conditions compared to many others.

Even though the number and naming of neighborhoods varied slightly between datasets, the consistent overlap revealed Lawrenceville as one of the best neighborhoods when balancing all three factors — low crime rate, high income, and good air quality.

### Conclusion

Although our datasets did not perfectly align in the number and naming of neighborhoods, the overlapping data consistently identified Lawrenceville as one of Pittsburgh’s stronger neighborhoods. It ranks within the top 30% for income, the top 35% for lowest crime rate, and the top 30% for air quality. These factors collectively suggest that Lawrenceville offers a balanced combination of safety, economic stability, and environmental quality.
